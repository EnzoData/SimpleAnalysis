Import the proper notebooks

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

Getting the Data and Shape

```python
df = pd.read_csv("../input/nyc-jobs.csv")
df.shape
```

Looking to see if we we have missing values

```python
df.count().sort_values()
```

'Recruitment contact' is empty and 'Post Until' is not useful for the EDA. We can drop them

```python
df.drop(['Recruitment Contact', 'Post Until'], axis=1, inplace=True)
```

Let's check some of the most important categorical features

```python
sns.countplot(x = "Full-Time/Part-Time indicator", data = df, palette = "Set3")
```

```python
sns.countplot(x = "Salary Frequency" , data = df, palette = "Set3")
```

```python
sns.countplot(x = "Level" , data = df, order = df['Level'].value_counts().index, palette = "Set3")
```

```python
plt.figure(figsize=(10,10))
plot = sns.countplot(y = "Job Category", data = df, order = df['Job Category'].value_counts().iloc[:10].index, palette = "Set3")
```

Some variables's distribution

```python
sns.distplot(df['Salary Range From'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
```

```python
sns.distplot(df['Salary Range To'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
```

```python
sns.distplot(df['# Of Positions'], 
             kde_kws={"color": "coral", "lw": 1, "label": "KDE"}, 
             hist_kws={"histtype": "stepfilled", "linewidth": 1, "alpha": 1, "color": "skyblue"});
```
